# 3. Data Cleaning

## Check Null Values

```python
transaction.isnull().sum()
```

## Card Present Flag

```python
transaction['card_present_flag'].value_counts(normalize = True)
```

As 80% of cards are flagged. I will fill replace the null value with `1`

```python
transaction['card_present_flag'] = transaction['card_present_flag'].fillna(1)
```

## Bpay Biller Code

```python
transaction[~transaction['bpay_biller_code'].isnull() & transaction['bpay_biller_code'] != 0]
```

As `0` covers the majority of row, I will replace the null value with `0`

```python
transaction['bpay_biller_code'] = transaction['bpay_biller_code'].fillna(0)
```

## Merchant State

```python
transaction['merchant_state'].value_counts(normalize = True)
```

As `NSW` covers the majority of the dataset population, I will replace the null values with `NSW`

```python
transaction['merchant_state'] = transaction['merchant_state'].fillna(value = 'NSW')
transaction['merchant_state'].isna().sum()
```

## Merchant Suburb

```python
transaction['merchant_suburb'] = transaction['merchant_suburb'].fillna(value = 'Sydney')
transaction['merchant_suburb'].isna().sum()
```

## Merchant Longtitude and Langtitude

```python
transaction['merchant_long_lat']  = transaction['merchant_long_lat'].fillna('151.21 -33.87')
transaction['merchant_long_lat'].isnull().sum()
```

## 7. Prediction Analysis

```python
# Import Linear Regression

from sklearn import linear_model
from sklearn.model_selection import train_test_split
import statsmodels.api as sm
```

```python
transaction['txn_description'].unique()
```

```python
# Subset the data where transaction description is salary

salary = transaction[transaction['txn_description'] == 'PAY/SALARY']
salary.shape
```

```python
drop_columns = ['status',
               'card_present_flag',
               'first_name',
               'extraction',
               'merchant_suburb',
               'merchant_state',
               'extraction',
               'country',
               'customer_id',
               'merchant_long_lat',
               'month',
               'year',
               'day_week',
               'week_number']

salary.drop(columns = drop_columns, inplace = True)
```

```python
# Correlation Heat Maps

plt.figure(figsize = (10,10))
c = salary.corr()
sns.heatmap(c, cmap = 'BrBG', annot = True)
c
```

### Linear Regression

```python
# import the mean squared error
from sklearn.metrics import mean_squared_error

features = ['age', 'balance']
x = salary[features]
y = salary['amount']

# Define the Linear Regression
reg = linear_model.LinearRegression()

# Split into train & test
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.33, random_state = 42)

# Fit the model
reg.fit(x_train,y_train)

# predict
y_pred = reg.predict(x_test)

# calculate the mean squared error
print('Mean Squared Error for Linear Regression: ', mean_squared_error(y_test, y_pred))
print('model score: ', reg.score(x_test, y_test))
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor

features = ['age', 'balance']
x = salary[features]
y = salary['amount']

# Define the Decision Tree Regressor
dtr = DecisionTreeRegressor()

# Split into train & test
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.33, random_state = 42)

# Fit the model
dtr.fit(x_train, y_train)

# predict price
y_pred = dtr.predict(x_test)

# calculate the mean squared error and score
print('Mean Squared Error for Decision Tree Regression: ', mean_squared_error(y_test, y_pred))
print('model score: ', dtr.score(x_test, y_test))
```

## Conclusion:

In conclusion, the simple linear regression is not very accurate with model score and mean squared error of 0.14011442295677012 and 1387994.7716570988 respectively.
